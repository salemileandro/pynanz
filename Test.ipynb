```python
def download(tickers, start, end, remove_nan = True):
    data = []
    for ticker in tickers:
        x = yf.Ticker(ticker)
        df = yf.download(ticker, start=start, end=end, interval="1d")
        if isinstance(df, pd.DataFrame):
            if len(df) == 0:
                continue
            if "Dividends" in df.columns:
                df.drop(columns=["Dividends"], inplace=True)
            if "Stock Splits" in df.columns:
                df.drop(columns=["Stock Splits"], inplace=True)
            new_cols = pd.MultiIndex.from_tuples([(ticker, c.lower()) for c in df.columns])
            df.columns = new_cols
            data.append(df)
            
    df = pd.concat(data, axis = 1)
    if remove_nan:
        df.dropna(axis=0, inplace=True)
        
    return df


df = download(["AAPL", "GOOG", "AIR.PA", "BTC-USD"], start="1980-01-01", end="2020-01-01")    
```
